```python
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import MDAnalysis as mda

saveplot = ".../results_plots/"
```

```python
mydir = '.../pmf-s8/US_50ns/WHAM/'
os.chdir(mydir)

profiles = [
    "profile_0.xvg", "profile_5000.xvg", "profile_10000.xvg", 
    "profile_15000.xvg", "profile_20000.xvg", "profile_25000.xvg", 
    "profile_30000.xvg", "profile_35000.xvg", "profile_40000.xvg", 
    "profile_45000.xvg"
]

plt.figure(figsize=(10, 5))
all_free_energy = []

for i, pro in enumerate(profiles):
    # Load the PMF data
    data = np.loadtxt(pro, comments=['@', '#', '&'])
    reaction_coordinate = data[:, 0]
    free_energy = data[:, 1]
    
    all_free_energy.append(free_energy)
    plt.plot(reaction_coordinate, free_energy, linewidth=4, alpha=0.07, color='blue')

    mirrored_rc = -reaction_coordinate[::-1] 
    mirrored_fe = free_energy[::-1]          
    plt.plot(mirrored_rc, mirrored_fe, linewidth=4, alpha=0.07, color='blue')

avg_free_energy = np.mean(all_free_energy, axis=0)
plt.plot(reaction_coordinate, avg_free_energy, color='black', linewidth=2, linestyle='--', label='PMF average')
mirrored_avg_fe = avg_free_energy[::-1]  
plt.plot(mirrored_rc, mirrored_avg_fe, color='black', linewidth=2, linestyle='--')


min_index = np.argmin(avg_free_energy)
max_index = np.argmax(avg_free_energy)

barrier_height = avg_free_energy[max_index] - avg_free_energy[min_index]
print(f"Energy barrier: {barrier_height:.2f} kJ/mol")
x_position = -4  # X position for the vertical line
y_min = avg_free_energy[min_index]
y_max = avg_free_energy[max_index]

plt.plot([x_position, x_position], [y_min, y_max], color='green', linestyle='--', linewidth=2)

plt.hlines(y=0, xmin=x_position-0.15, xmax=x_position + 0.15, color='green', linestyle='--', linewidth=2)
plt.hlines(y=y_min, xmin=x_position-0.15, xmax=x_position + 0.15, color='green', linestyle='--', linewidth=2)

plt.text(x_position + 0.1, (y_min + y_max) / 2, f'{barrier_height:.2f} kJ/mol', color='black', fontsize=14)

# Plotting settings
plt.xlabel('Reaction coordinate', fontsize=14)
plt.ylabel(r'Free energy profile (kJmol$^{-1}$)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.1)
plt.legend()
plt.tight_layout()
plt.tick_params(axis='both', labelsize=12)
plt.xticks(np.arange(-5., 5., 1))
plt.title("Grappa after retraining")
# plt.show()
plt.savefig(saveplot+'pmf-s8.pdf', format='pdf', bbox_inches='tight')
plt.savefig(saveplot+'pmf-s8.svg', format='svg', bbox_inches='tight')
```
